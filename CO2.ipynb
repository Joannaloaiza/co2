## Literature Review

The prolonged and consistent use of face masks, particularly in environments where carbon dioxide ($CO_{2}$) concentrations can reach elevated levels, has raised concerns about potential health implications. To address this multifaceted issue, it is essential to consider various contexts, including indoor settings such as offices, where $CO_{2}$ concentrations can fluctuate significantly, surpassing established safety standards [[5]](#References).

The concentration of $CO_{2}$ in indoor environments is influenced by factors like occupancy, room size, and ventilation. While outdoor air typically contains approximately 400 parts per million (ppm) of $CO_{2}$, indoor levels can vary widely, reaching levels well above this baseline [[5]](#References). Excessive $CO_{2}$ concentrations within indoor spaces can lead to a range of symptoms, including headaches, dizziness, fatigue, and difficulties in concentration . The severity of these symptoms often correlates with the degree of $CO_{2}$ elevation.

A seminal study by Satish and Mendell [[6]](#References) examined the impact of varying $CO_{2}$ concentrations on cognitive abilities, shedding light on the potential risks associated with high $CO_{2}$ exposure. The study involved 22 individuals divided into six groups, each exposed to different $CO_{2}$ concentrations: 600 ppm, 1000 ppm, and 2500 ppm. Cognitive assessments were conducted at the conclusion of each experiment, with noteworthy results. Notably, the cognitive performance of participants exposed to 600 ppm and 1000 ppm of $CO_{2}$ displayed similar outcomes. However, in the case of the 2500 ppm concentration, decision-making performance exhibited a significant decline. This finding underscores the potential cognitive effects of elevated $CO_{2}$ levels.

Concerns have arisen regarding the impact of mask-wearing, particularly in environments where $CO_{2}$ concentrations may be elevated. Masks, by design, create a barrier that can trap exhaled air, potentially leading to a slight increase in $CO_{2}$ concentration within the mask. However, it is crucial to emphasize that masks are engineered to permit the exchange of air, and under typical conditions, $CO_{2}$ levels within the mask should not reach harmful thresholds during regular usage. Therefore, the risk of adverse health effects directly attributed to $CO_{2}$ buildup within masks is generally considered low in standard indoor environments [[6]](#References).

Nevertheless, it is essential to strike a balance between safety and comfort. For most individuals, the benefits of mask-wearing, including the mitigation of disease transmission, outweigh potential discomfort. Moreover, individual tolerance to elevated $CO_{2}$ levels may vary, with some individuals being more sensitive to changes in air quality and potentially experiencing discomfort or cognitive effects at lower $CO_{2}$ concentrations [6].

Azuma and Osawa [[3]](#References) underscore the direct effects of low-level carbon dioxide ($CO_{2}$) exposure on human health. To comprehensively address the implications of elevated $CO_{2}$ levels, it is essential to consider not only $CO_{2}$ but also other gases commonly encountered in indoor environments. Key guidelines and standards are established to safeguard individuals' health in these settings:

$CO_{2}$ Ceiling Concentration: The U.S. Department of Health, Education, and Welfare stipulate a ceiling concentration for carbon monoxide (CO), a related gas, of 200 parts per million (ppm) in the work environment [[2]](#References). This emphasizes the importance of establishing and adhering to safety standards concerning air quality within indoor spaces to protect workers' well-being.

OSHA PEL: The Occupational Safety and Health Administration (OSHA) in the United States sets a permissible exposure limit (PEL) for carbon monoxide at 50 ppm parts of air (55 milligrams per cubic meter (mg/m³)) as an 8-hour time-weighted average (TWA) concentration [[4]](#References). This regulatory standard highlights the significance of controlling and monitoring indoor air quality, with specific attention to CO levels [[4]](#References).

The U.S. Department of Health, Education, and Welfare's ceiling concentration for CO and OSHA's PEL for CO reinforce the importance of maintaining safe indoor air quality. While these standards primarily address CO, they underscore the broader concern of indoor air quality management, which encompasses gases like $CO_{2}$   [[2] [4] [6]](#References).


To mitigate concerns associated with elevated $CO_{2}$ levels in indoor settings, particularly in shared spaces like offices, adequate ventilation plays a critical role. Proper ventilation ensures the maintenance of indoor air quality and reduces the risk of elevated $CO_{2}$ levels, thus contributing to the comfort and well-being of occupants. Monitoring and addressing elevated $CO_{2}$ concentrations, particularly when they approach or exceed levels like 2500 ppm, as demonstrated in the study by Satish and Mendell [6], should remain a priority in indoor air quality management.

Exposure to CO2 can produce a variety of health effects. These may include headaches, dizziness, restlessness, a tingling or pins or needles feeling, difficulty breathing, sweating, tiredness, increased heart rate, elevated blood pressure, coma, asphyxia, and convulsions.
The levels of CO2 in the air and potential health problems are:
•	400 ppm: average outdoor air level.
•	400–1,000 ppm: typical level found in occupied spaces with good air exchange.
•	1,000–2,000 ppm: level associated with complaints of drowsiness and poor air.
•	2,000–5,000 ppm: level associated with headaches, sleepiness, and stagnant, stale, stuffy air. Poor concentration, loss of attention, increased heart rate and slight nausea may also be present.
•	5,000 ppm: this indicates unusual air conditions where high levels of other gases could also be present. Toxicity or oxygen deprivation could occur. This is the permissible exposure limit for daily workplace exposures.
•	40,000 ppm: this level is immediately harmful due to oxygen deprivation.

**Telaire T6703 CO2 sensor**
this device utilizes non-dispersive infrared (NDIR) technology for measuring carbon dioxide (CO2) levels. The sensor employs the following steps:

1. Infrared Light Source: The sensor incorporates an infrared (IR) light source, typically an IR LED.
2. Gas Sample Chamber: A chamber is provided for introducing air containing CO2.
3. Absorption of Infrared Light: Infrared light, emitted by the source, passes through the gas chamber. CO2 molecules absorb specific infrared wavelengths.
4. Detection of Absorbed Light: An IR detector measures the intensity of transmitted light through the sample.
5. CO2 Concentration Calculation: The absorbed infrared light's amount correlates with CO2 concentration. The sensor's electronics analyze this data to calculate CO2 levels.
6. Output Signal: The sensor generates an output signal proportional to the CO2 concentration. Output can be in analog voltage, digital (UART or I2C), or other formats, depending on the sensor model

**Event Generation**

The frequency of events associated with a timer is generated according to the following expression: Event = TIM_CLK/((PSC + 1)(ARR + 1)(RCR + 1)) with:
- PSC, Pre Scalar.
- ARR, Auto Reload Register.
- RCR, Repetition Counter Register.

In many CubeMX configurations, RCR does not appear, so it is assigned a value of zero. Refer to AN4013 section 2.2 for different configuration strategies. The project's configuration is as follows:
- TIM_CLK de 16 MHz.
- PSC de 16-1.
- ARR de 65535-1.
- RCR de 0.

As a result, an event is generated every 15.25 Hz, which is equivalent to 0.06557 seconds. See the following calculation:

Event = (16 * 10^6) Hz / ( ( 16 - 1 + 1) * ( 65535-1 + 1) * (0 + 1))
Event = 16,000,000 Hz / 1,048,560
Event = 15.25 Hz

Consider 5 time units, 16 MHz, and 23 microseconds. One time unit corresponds to:

1 time unit = 23 microseconds / 5 units = 4.6 microseconds per unit.

Now, the amount of time per clock tick associated with timer TIM6 will be Time per tick = 1 / clock frequency, which results in:

Time per tick = 1 / 15.25 Hz ≈ 0.06557377049 seconds

This is equivalent to 0.06557377049 seconds * 1,000,000 microseconds/second ≈ 65,573.77 microseconds. This is the approximate time increment for each clock tick. However, time increments are not constant. Therefore, we assume a linear trend in the time increment and estimate the parameters for delay calculation as follows:

With "y" as the delay in microseconds and "x" as time units.
